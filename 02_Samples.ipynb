# Modelo de Classificação de Animais - Cachorro ou Porco

```python
# featyres (1 sim, 0 não)
# pelo longo?   1 - longo   e 0 - curto
# perna curta?  1 - curta   e 0 - longa
# faz au au?    1 - late    e 0 - não late

## Dados de treinamento
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]
```

```python
dados = [porco1,
         porco2,
         porco3,
         cachorro1,
         cachorro2,
         cachorro3]

# 1 => porco, 0 => cachorro
classes = [1, 1, 1, 0, 0, 0] 
```

```python
dados
```

## Criar e treinar o modelo

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
```

```python
model.fit(dados, classes)
```

## Teste de Predição/Classificação dos Dados

```python
# Necessário preparar uma matriz (tabela) com um único valor
animal_misterioro = [[1, 1, 1],
                     [0, 0, 0],
                     [0, 0, 1]]
```

```python
# A função predict espera uma matriz de dados
previsoes = model.predict(animal_misterioro)
print(previsoes)

# resultados reais = [0, 1, 1]
# resultados foi = [0, 1, 1]
```

## Avaliações

```python
# resultados reais = [0, 1, 1]
# resultados foi = [0, 1, 1]

testes = [[1, 1, 1], [0, 0, 0], [0, 0, 1]]
previsoes = model.predict(testes)
testes_classes = [0, 1, 1]
print(previsoes)
```

```python
corretos = (previsoes == testes_classes).sum()
total = len(testes)
taxa_de_acerto = corretos/total
print("Taxa de acerto", taxa_de_acerto * 100)
```

```python
taxa_erro = 1 - taxa_de_acerto
print("Taxa de Erro: ", taxa_erro * 100)
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto = accuracy_score(testes_classes, previsoes)
taxa_acerto
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(testes_classes, previsoes))
```
